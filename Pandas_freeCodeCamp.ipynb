# Series

```python
import pandas as pd

# Create a Series
data = [10, 20, 30, 40, 50]
series = pd.Series(data)

# Create a string Series
string_data = ['apple', 'banana', 'cherry', 'date', 'elderberry']
string_series = pd.Series(string_data)

# Display the Series
print("Series:")
print("---------------------------------------")
display(series)
print("---------------------------------------")
display(string_series)
print("---------------------------------------")
certificates_earned = pd.Series([8, 2, 5, 6], index=['Tom', 'Kris', 'Ahmad', 'Beau'])
display(certificates_earned)
print("---------------------------------------")

# Operation methods
print("Accessing element of Series :")
print("First element :",certificates_earned.iloc[0]) # Indexing
print("Last element :",certificates_earned.iloc[-1]) # Indexing
print("First element :",certificates_earned.loc['Tom']) # Labeling
print("Last element :",certificates_earned.loc['Beau']) # Labeling
print("From 0 to 3 (excluded) elements :",certificates_earned.iloc[0:3])
print("---------------------------------------")
print("Converting element of Series :")
series = series.astype(float)
print(series)
print("---------------------------------------")
print("Conditional selection in Series :")
print("Certified earned above 5 :")
print(certificates_earned > 5) # True or False condition
print(certificates_earned[certificates_earned > 5]) # select rows where condition is verified
print("---------------------------------------")
print("Certified earned between 5 and 10 :")
print(certificates_earned[(certificates_earned > 5) & (certificates_earned < 10)]) # select rows where condition is verified
print("---------------------------------------")
print("Certified earned less than 5 or more than 10 :")
print(certificates_earned[(certificates_earned > 5) | (certificates_earned < 10)]) # select rows where condition is verified
print("---------------------------------------")
```

# Pandas dataFrames

```python
import pandas as pd


# Create a dataframe
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']
       }

df = pd.DataFrame(data)
display(df)
print("---------------------------------------")
print("Selection :")
# Using iloc
print("Row at index 0:")
display(df.iloc[0])
print("\nRows at index 0 and 2:")
display(df.iloc[[0, 2]])
print("\nRows from index 1 to 3 (exclusive):")
display(df.iloc[1:3])
print("\nCell at row 1, column 2:")
display(df.iloc[1, 2])
print("\nSlice of rows and columns:")
display(df.iloc[0:2, 0:2])

print("---------------------------------------")
# Using loc
print("Row with index label 0:")
display(df.loc[0])
print("\nRows with index labels 0 and 2:")
display(df.loc[[0, 2]])
print("\nRows from index label 1 to 3 (inclusive):")
display(df.loc[1:3])
print("\nCell at row with index label 1, column 'Age':")
display(df.loc[1, 'Age'])
print("\nSlice of rows and columns using labels:")
display(df.loc[0:2, ['Name', 'City']])

# Conditional selection
print("---------------------------------------")
print("Conditional selection :")
print("Rows where Age is greater than 30:")
display(df[df['Age'] > 30])
print("Rows where City is 'New York':")
display(df[df['City'] == 'New York'])

# Modifying dataframe
print("---------------------------------------")
print("Modifying dataframe :")
df.loc[2, 'Age'] = 36 # Replacing values
display(df)
print("---------------------------------------")
df['Salary'] = [50000, 60000, 70000, 80000] # Adding new column
display(df)
print("---------------------------------------")
# Delete column
df.drop(columns=['Salary'], inplace=True)
display(df)
print("---------------------------------------")
# Rename column
df.rename(columns={'Name': 'First Name'}, inplace=True)
display(df)
print("---------------------------------------")
# Adding Last Name column
df['Last Name'] = ['Smith', 'Johnson', 'Williams', 'Brown']
display(df)
print("---------------------------------------")
# Concatenate columns
df['Full Name'] = df['First Name'] + ' ' + df['Last Name']
display(df)
print("---------------------------------------")
# Arrange columns
df = df[['First Name', 'Last Name', 'Full Name', 'Age', 'City']]
display(df)
print("---------------------------------------")
# Using apply function
df['Age'] = df['Age'].apply(lambda x: x + 1)
display(df)
print("---------------------------------------")
# Use map function (add new column called city ancronym)
city_map = {'New York': 'NY', 'San Francisco': 'SF', 'Los Angeles': 'LA', 'Chicago': 'CH'}
df['City Abbreviation'] = df['City'].map(city_map)
display(df)
print("---------------------------------------")
# Add country column
df['Country'] = 'USA'
display(df)
print("---------------------------------------")
```

# Data cleaning
#### https://www.geeksforgeeks.org/data-analysis/working-with-missing-data-in-pandas/
#### https://www.geeksforgeeks.org/python/count-nan-or-missing-values-in-pandas-dataframe/

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Handling missing data
df = pd.DataFrame({'A': [1, 2, None, 4],
                   'B': [5, None, 7, 8],
                   'C': [9, 10, 11, None]})

display(df)
print("---------------------------------------")
print("Detect missing values :")
# Checking Missing Values in Pandas
print(df.isnull())
print("Total missing values :",df.isnull().sum().sum()) # Checking the amount of them
#print(df.isna())
#print(df.notnull())
print("---------------------------------------")
# Drop rows with missing values
print("Drop rows with missing data :")
df_dropped = df.dropna()
display(df_dropped)
print("---------------------------------------")
# Fill missing values with a specific value
print("Fill missing values with a specific value :")
df_filled = df.fillna(0)
display(df_filled)
print("---------------------------------------")
# Fill missing values with the mean of the column
print("Fill missing values with the mean of the column :")
# Round the mean pleas
df_filled_mean = df.fillna(df.mean().round(2))
display(df_filled_mean)
print("---------------------------------------")
# Suppose our dataframe is cleaned from duplicatee, visualize the numeric values
print("Visualize the numeric values with scatter subplots:")
sns.pairplot(df)
plt.show()
print("---------------------------------------")
```

```python
import csv

# https://www.geeksforgeeks.org/python/writing-csv-files-in-python/

# Data to be written to the CSV file
header = ['Organization', 'CEO', 'Established']
data = [
    ['Alphabet', 'Sundar Pichai', '02-Oct-15'],
    ['Microsoft', 'Satya Nadella', '04-Apr-75'],
    ['Amazon', 'Jeff Bezos', '05-Jul-94']
]

# Open the file in write mode
with open('organizations.csv', 'w', newline='') as file:
    writer = csv.writer(file)

    # Write the header
    writer.writerow(header)

    # Write the data rows
    writer.writerows(data)

print("organizations.csv has been created successfully.")
```
